```python
import tensorflow as tf
```

```python
filenames = [
    'data/fake_rings.tfrecord'
]
raw_dataset = tf.data.TFRecordDataset(filenames)
```

```python
N_RINGS = 100
parse_schema = {
    'id': tf.io.FixedLenFeature([], tf.int64),
    'et': tf.io.FixedLenFeature([], tf.float32),
    'eta': tf.io.FixedLenFeature([], tf.float32),
    'avgmu': tf.io.FixedLenFeature([], tf.float32),
    'label': tf.io.FixedLenFeature([], tf.int64),
}
for i in range(N_RINGS):
    parse_schema[f'ring_{i}'] = tf.io.FixedLenFeature([], tf.float32)

def parse_example(serialized_string):
    return tf.io.parse_single_example(
        serialized_string,
        parse_schema
    )
```

```python
parsed_dataset = raw_dataset.map(parse_example)
```

```python
for i, example in enumerate(parsed_dataset.take(3)):
    print(f"Example {i}: {type(example)}")
    print(example)
```
