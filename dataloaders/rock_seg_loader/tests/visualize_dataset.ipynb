```python
#Para utilizar com um arquivo rock_dataloader.py
from rock_dataset_multi import RockSegmentationDatasetMulti
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt



dataset_path = r"D:\LRAP_DOUTORADO\4_CPE883_ML2_2p\2_Dataset"  # ou caminho completo, se estiver fora
dataset = RockSegmentationDatasetMulti(root_dir=dataset_path, output_shape=(512, 512))
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

for images, masks, amostras in dataloader:
    for i in range(len(images)):
        print(f"Amostra: {amostras[i]} — Shape: {images[i].shape}, Máscara: {masks[i].shape}")

        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(images[i][0], cmap='gray')
        plt.title(f"Imagem ({amostras[i]})")
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(masks[i], cmap='viridis', vmin=0, vmax=2)
        plt.title("Máscara")
        plt.axis('off')
        plt.show()
    break
```
